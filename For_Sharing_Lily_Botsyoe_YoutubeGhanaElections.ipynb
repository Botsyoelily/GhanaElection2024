```python
from googleapiclient.discovery import build
from googleapiclient.errors import HttpError
```

```python
#Removed API key, generate yours for use.
api_key= "ADD_YOUR_API_HERE"
api_name= "youtube"
api_version= "v3"
youtube= build(api_name,api_version,developerKey=api_key)
```

```python
# Initialize variables
query = "John Dramani Mahama 2024"  # Search query for the NDC presidential candidate
max_results_per_page = 50  # Maximum allowed results per API request
total_videos = 100  # Desired number of videos
retrieved_videos = 0  # Counter for retrieved videos
video_ids = []  # List to store video IDs
next_page_token = None  # Token for pagination

# Iterate to fetch up to 100 videos
while retrieved_videos < total_videos:
    # Executes a search request
    search_response = youtube.search().list(
        part="snippet",
        q=query,
        type="video",
        maxResults=max_results_per_page,
        pageToken=next_page_token
    ).execute()

    # Extract video IDs from the search results
    video_ids.extend([item['id']['videoId'] for item in search_response.get('items', [])])
    retrieved_videos = len(video_ids)

    # Check for more pages
    next_page_token = search_response.get('nextPageToken')
    if not next_page_token:
        break

# Limit to exactly 100 video IDs (if more were fetched)
video_ids = video_ids[:total_videos]

# If video IDs were found, retrieve statistics
if video_ids:
    # Split video IDs into chunks of 50 (API limit for statistics)
    for i in range(0, len(video_ids), max_results_per_page):
        chunk = video_ids[i:i + max_results_per_page]
        statistics_response = youtube.videos().list(
            part="statistics",
            id=','.join(chunk)
        ).execute()

        # Process and display video statistics
        for video in statistics_response.get('items', []):
            video_id = video['id']
            stats = video['statistics']
            print(f"Video ID: {video_id}")
            print(f"Statistics: {stats}")
```

Step 2: Extract Comments and Perform Sentiment Analysis

```python
from textblob import TextBlob

# Function to classify sentiment
def get_sentiment(text):
    analysis = TextBlob(text)
    if analysis.sentiment.polarity > 0:
        return "positive"
    elif analysis.sentiment.polarity < 0:
        return "negative"
    else:
        return "neutral"
```

```python
!pip install --upgrade google-api-python-client google-auth-httplib2 google-auth-oauthlib
#Install the necessary google api client library

from googleapiclient.discovery import build
from googleapiclient.errors import HttpError #Import the HttpError class
from textblob import TextBlob
#Import necessary libraries

# ... (Your existing code to set up the YouTube API client) ...


def analyze_comments(video_id):
    comments_data = []
    try:
        comments_response = youtube.commentThreads().list(part="snippet", videoId=video_id, maxResults=20).execute()

        for item in comments_response["items"]:
            comment_text = item["snippet"]["topLevelComment"]["snippet"]["textDisplay"]
            sentiment = get_sentiment(comment_text)
            comments_data.append({"main_text": comment_text, "sentiment": sentiment})

    except HttpError as e: #Change exception to HttpError
        if e.resp.status == 403:
            print(f"Comments are disabled for video ID: {video_id}")
        else:
            print(f"An error occurred: {e}")

    return comments_data
```

Step 3: Aggregate Sentiment Results

```python
bawumia_scores = {"positive": 0, "neutral": 0, "negative": 0}
mahama_scores = {"positive": 0, "neutral": 0, "negative": 0}
```

```python
# Function to update scores
def update_scores(scores, comments):
    for comment in comments:
        scores[comment["sentiment"]] += 1
```

```python
# Aggregate scores for John Dramani Mahama videos
for video in search_response["items"]:
    comments = analyze_comments(video["id"]["videoId"])
    update_scores(mahama_scores, comments)
```

```python
# Aggregate scores for Mahamudu Bawumia videos
for video in bawumia_response["items"]:
    comments = analyze_comments(video["id"]["videoId"])
    update_scores(bawumia_scores, comments)
```

Explanation of Determination
The winner is projected by comparing the aggregate sentiment score for each candidate. This method assumes that the candidate with a more positive sentiment (higher score) has greater public favor in the sample data.

This approach allows you to analyze and interpret public sentiment towards each candidate and make a creative prediction based on the data.

```python
# Aggregate scores for John Dramani Mahama videos
for video in search_response["items"]:
    comments = analyze_comments(video["id"]["videoId"])
    update_scores(mahama_scores, comments)

# Aggregate scores for Mahamudu Bawumia videos
for video in bawumia_response["items"]:
    comments = analyze_comments(video["id"]["videoId"])
    update_scores(bawumia_scores, comments)

# Calculate total scores by summing the values in the dictionaries
mahama_total = sum(mahama_scores.values())
bawumia_total = sum(bawumia_scores.values())

```

```python
# Determine winner
if mahama_total > bawumia_total:
    print("Based on sentiment analysis, John Dramani Mahama is projected to win.")
else:
    print("Based on sentiment analysis, Mahamudu Bawumia is projected to win.")
```
